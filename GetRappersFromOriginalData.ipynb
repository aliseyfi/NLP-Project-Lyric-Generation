```python
import pandas as pd
```

```python
df = pd.read_csv('lyrics.csv')
```

```python
df['dummy'] = 1
```

```python
grouped_df = df.groupby(df['artist']).dummy.sum().reset_index().sort_values("dummy", ascending=False)
```

```python
drake_df = df.loc[df['artist'] == 'drake']
```

```python
# Get dataframe of select artists
artists = ['drake', 'chris-brown', 'eminem', '50-cent', '2pac', 'e-40', 'chief-keef']
final_df = df[df['artist'].isin(artists)]
```

```python
final_df = final_df.drop('dummy', axis=1)
```
